## 作業
請找出一個你認為機器學習可以發光發熱的領域並闡述原因? 

【答】現在機器學習較夯的領域是在語音辨識、影像辨識以及自然語言處理，我認為應用上醫療是最直接可發光發熱的領域，因為在台灣健保制度周全的優勢狀態，累績的資料量絕對足夠發展人工智慧，而其他領域例如自然災害的預測則可能較為劣勢的，地震、海嘯發生次數太過於少相較於正常狀態，因此會有資料類別不均勻的問題，當然醫療上也有這問題存在，例如罕見疾病的預測都是一大困難。但近期也提出許多演算法可自動生成少數類別的資料，例如：過取樣的SMOTE等，但生成的資料相較於實際資料來說其真實性較低。

我以下提出幾個例子，是目前我從網上查詢或各期刊論文閱讀中得知人工智慧已應用的領域，並統整後提出一些衍伸想法：

在醫療領域上語音辨識目前有聽診器聽出心音來辨別是否有心臟疾病；影像辨識可用於超音波、X光照、眼底鏡等對人體的內部結構都可作是否罹患疾病的判別；但自然語言處理在醫療的應用上則較難完全展現，或許可透過爬蟲得知在各網站收集醫療服務體驗，來得知病患的需求滿意度、醫療流程的順暢度，改善醫療服務品質。而在3C產品人工智慧需求也漸漸增加，例如：手機相機的影像拍攝，擷取拍下那瞬間，從各時間段的影像選擇高品質的作存檔、利用人工智慧修圖，根據累積的老人、小孩臉部狀態來重新組合被拍攝者的臉部面容、存取使用者常用的習慣，累積資料推測使用者的喜好。工業上取代對於高頻繁性的工作(人工檢查產品瑕疵)、金融業股市的預測等。遊戲與生活品質提升則有老人居家照顧、無人商店、AlphaGo遊戲上的應用、自動駕駛汽車、AR、VR等。

整體而言，不管是用於哪個領域，我認為資料的收集是非常重要的，資料的公開化是有助於人工智慧的發展，但錯誤的資料可能使模型學習過程中導致預測的誤判，資料的量也是人工智慧成功的一項因素，當然現在仍有需多無監督式學習方法，自動標註方式來產生大量資料。

人的想法是無限的，各應用領域都有發展的潛能。

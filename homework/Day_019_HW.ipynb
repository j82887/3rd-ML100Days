- 如何依照不同的 HOUSETYPE_MODE 類別指定資料, 並繪製長條圖(.hist())? (In[3])

```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版

# unique函數能去除其中重複的元素，並按照元素由大到小返回一個新的無元素重複的列表
unique_house_type = app_train['HOUSETYPE_MODE'].unique()

#列表的長度個數，為要繪製的列個數
nrows = len(unique_house_type)

#繪製的欄個數
ncols = nrows // 2

#設定繪製字體
plt.rc('font',family='Times New Roman')

#繪製的大小(為整個視窗大小)
plt.figure(figsize=(18,30))
for i in range(len(unique_house_type)):
    
    #繪製擺放位置
    plt.subplot(nrows, ncols, i+1)
    
    #繪製直方圖，針對不同的'HOUSETYPE_MODE'對'AMT_CREDIT'，設定長條數量為50，邊線設定為有(k)
    app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]),'AMT_CREDIT'].hist(edgecolor = 'k', bins = 50)
    
    #繪製的標題
    plt.title(str(unique_house_type[i]), size =18)
    
    #繪製軸線字的大小
    plt.xticks(size = 14)
    plt.yticks(size = 14)

#繪製呈現
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app, engine='python')
# app_train內的資料已為DataFrame
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案

#### 1.資料的 標題名稱

```python
#得到第一列的標題名稱
data_name = app_train.columns # get col index
print(data_name)
```

#### 2.顯示前n列的值(n不含標題列)

```python
app_train.head() #預設是5
```

```python
app_train.head(2) #指定是2
```

#### 3.顯示後n列的值

```python
app_train.tail() #預設是5
```

#### 4.資料的值轉成陣列

```python
#值轉成陣列
data_array = app_train.as_matrix()
print(data_array)
```

#### 5.每一欄資料的型態

```python
app_train.dtypes
```

#### 6.資料的 size 

```python
#欄數乘以列數(不包含標題列)

#利用DataFrame轉成array的方式
print("資料大小：%s" %data_array.size)
#利用DataFrame轉成array的方式
print("資料大小：%s" %app_train.size)
```

#### 7.資料的 row 數以及 column 數

```python
## ===== 利用len方式 ======

#利用DataFrame轉成array的方式
print("資料的列數量：%s" %(len(data_array))) #不含標題列
print("資料的欄位數量：%s" %(len(data_array.T))) #轉置後欄變為列，所以取列的數量
print("資料的欄位數量：%s" %(len(data_array[0]))) #取第一列的數量，即欄的數量

#利用DataFrame內的方式
print("資料的列數量：%s" %len(app_train.index))
print("資料的欄位數量：%s" %(len(app_train.head().T))) #轉置後欄變為列，所以取列的數量
print("資料的欄位數量：%s" %(len(app_train.columns)))  

(data_detail, data_name) = app_train.axes
print("資料的列數量：%s" %len(data_detail))
print("資料的欄位數量：%s" %(len(data_name)))

## ===== 利用shape方式 ======

#利用DataFrame轉成array的方式
print("資料的列數量：%s" %(data_array.shape[0]))
print("資料的欄位數量：%s" %(data_array.shape[1]))

#利用DataFrame內的方式
print("資料的列數量：%s" %(app_train.shape[0]))
print("資料的欄位數量：%s" %(app_train.shape[1]))
```

#### 8.列出所有欄位

```python
#上面已用.columns方式，呈現所有欄位

#轉成列表顯示所有欄位名稱
print("顯示所有標題名稱(列表)：\n", data_name.tolist())
print("\n")

#轉成陣列顯示出所有欄位名稱
print("顯示所有標題名稱(陣列)：\n", np.array(data_name))
```

#### 9.截取部分資料

```python
#提取第四列到第五列，且第三欄到第四欄的所有值 
app_train.iloc[4:6, 3:5] #4到6但不包含6，3到5但不包含5
```

```python
#若知道標題名稱，來提取第四列到第五列的值
print(app_train['CODE_GENDER'][4:6])
print(app_train['FLAG_OWN_CAR'][4:6])
```
